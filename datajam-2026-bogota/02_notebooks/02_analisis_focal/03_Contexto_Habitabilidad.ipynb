👉 Este análisis explica el 'contexto'. Identificaremos las localidades con peores indicadores y armaremos un mapa de vulnerabilidad general.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from pathlib import Path

sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

### Carga de Datos Preprocesados

```python
# Ruta robusta: fallback a ruta absoluta si la relativa no funciona
_candidates = [
    Path('../../01_datos/processed'),
    Path('../../../01_datos/processed'),
    Path(r'C:/Users/Latitude/Documents/V Semestre/DATAJAM/DATAJAM-DATAFONO_2026/datajam-2026-bogota/01_datos/processed'),
]
PROCESSED_DIR = next((p for p in _candidates if p.exists()), None)
assert PROCESSED_DIR is not None, f'No se encontro el directorio processed. CWD={Path.cwd()}'

file_path = PROCESSED_DIR / 'p_condiciones_habitabilidad.csv'
df_habi = pd.read_csv(file_path)
print(f'Total de registros cargados: {len(df_habi):,}')
display(df_habi.head(3))
```

## 📍 Territorial
**¿Qué localidades tienen peores indicadores? Ranking de vulnerabilidad**

```python
# ¿Qué localidades tienen peores indicadores? Ranking de vulnerabilidad
ranking_loc = (
    df_habi.groupby('localidad')['valor_indicador']
    .mean()
    .sort_values(ascending=False)
    .head(15)
    .reset_index()
)
ranking_loc.columns = ['localidad', 'valor_promedio']

plt.figure(figsize=(12,7))
sns.barplot(data=ranking_loc, x='valor_promedio', y='localidad', palette='magma')
plt.title('Promedio de Valor de Indicadores por Localidad')
plt.xlabel('Valor Promedio del Indicador')
plt.ylabel('Localidad')
plt.tight_layout()
plt.show()
```

## 📊 Indicadores clave
**¿Cuáles indicadores son más críticos y cómo varían entre zonas?**

```python
# ¿Cuáles indicadores son más críticos y cómo varían entre zonas?
top_indicadores = (
    df_habi['clase_indicador']
    .value_counts()
    .head(8)
    .reset_index()
)
top_indicadores.columns = ['clase_indicador', 'frecuencia']

plt.figure(figsize=(12,5))
sns.barplot(data=top_indicadores, x='frecuencia', y='clase_indicador', palette='viridis')
plt.title('Tipos de Indicadores Medidos con Frecuencia')
plt.xlabel('Frecuencia de Registro')
plt.ylabel('Clase de Indicador')
plt.tight_layout()
plt.show()
```

**¿Hay clusters de alta vulnerabilidad a nivel de UPZ?**

```python
# ¿Hay clusters de alta vulnerabilidad a nivel de UPZ?
ranking_upz = df_habi['unidad_planeacion_zonal'].value_counts().head(10)
print('Top 10 UPZs con mayor intensidad o frecuencia de fallas/indicadores:\n')
print(ranking_upz.to_string())
```
