# 💀 2. Salud: Suicidio (RESULTADO MÁS GRAVE)

👉 Este análisis sirve para validar si los patrones de ideación escalan a hechos consumados de forma proporcional.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set_theme(style='darkgrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

### Carga de Datos Preprocesados

```python
PROCESSED_DIR = '../../01_datos/processed'
file_path = os.path.join(PROCESSED_DIR, 'p_salud_suicidio.xlsx')
df_suicidio = pd.read_excel(file_path)
print(f'Total de registros cargados: {len(df_suicidio):,}')
display(df_suicidio.head())
```

## 📈 Tendencia
**¿También está aumentando al igual que la ideación?**

```python
col_ano = 'ANO_DEL_HECHO' if 'ANO_DEL_HECHO' in df_suicidio.columns else 'ano'
if col_ano in df_suicidio.columns:
    tendencia = df_suicidio[col_ano].value_counts().sort_index()
    plt.figure(figsize=(10,5))
    sns.lineplot(x=tendencia.index, y=tendencia.values, marker='s', color='crimson', linewidth=2)
    plt.title('Tendencia Histórica de Suicidios Consumados')
    plt.xlabel('Año')
    plt.ylabel('Cantidad de Casos')
    plt.xticks(tendencia.index)
    plt.show()
```

## 📍 Territorial
**¿Coincide con las mismas localidades de ideación?**

```python
col_loc = 'LOCALIDAD_DEL_HECHO' if 'LOCALIDAD_DEL_HECHO' in df_suicidio.columns else 'localidad'
if col_loc in df_suicidio.columns:
    loc_counts = df_suicidio[col_loc].value_counts().head(10)
    plt.figure(figsize=(12,6))
    sns.barplot(x=loc_counts.values, y=loc_counts.index, palette='mako')
    plt.title('Top Localidades con Mayor Registro de Suicidios')
    plt.xlabel('Casos Consumados')
    plt.ylabel('Localidad')
    plt.show()
```

## 👥 Perfil
**¿Mismo grupo etario? ¿Mismo sexo predominante?**

```python
col_edad = 'GRUPO_DE_EDAD_QUINQUENAL_' if 'GRUPO_DE_EDAD_QUINQUENAL_' in df_suicidio.columns else 'edad'
col_sexo = 'SEXO_DE_LA_VICTIMA' if 'SEXO_DE_LA_VICTIMA' in df_suicidio.columns else 'sexo'

if col_edad in df_suicidio.columns and col_sexo in df_suicidio.columns:
    plt.figure(figsize=(12,7))
    sns.countplot(data=df_suicidio, y=col_edad, hue=col_sexo, order=df_suicidio[col_edad].value_counts().index)
    plt.title('Perfil Demográfico: Edad y Sexo en Suicidios')
    plt.xlabel('Número de Casos')
    plt.ylabel('Grupo de Edad')
    plt.legend(title='Sexo')
    plt.show()
```

## ⚠️ Causas
**¿Qué razones o motivos de suicidio aparecen más?**

```python
col_razon = 'RAZON_DEL_SUICIDIO' if 'RAZON_DEL_SUICIDIO' in df_suicidio.columns else 'causa'
if col_razon in df_suicidio.columns:
    razones = df_suicidio[col_razon].value_counts().head(10)
    plt.figure(figsize=(12,6))
    sns.barplot(x=razones.values, y=razones.index, palette='flare')
    plt.title('Principales Razones Reportadas de Suicidio')
    plt.xlabel('Frecuencia')
    plt.ylabel('Motivo')
    plt.show()
```
