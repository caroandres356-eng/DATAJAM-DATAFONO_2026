# 🏠 3. Habitabilidad / Accesibilidad (FACTOR TERRITORIAL)

👉 Este análisis explica el 'contexto'. Identificaremos las localidades con peores indicadores y armaremos un mapa de vulnerabilidad general.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

### Carga de Datos Preprocesados

```python
PROCESSED_DIR = '../../01_datos/processed'
file_path = os.path.join(PROCESSED_DIR, 'p_condiciones_habitabilidad.xlsx')

if os.path.exists(file_path):
    # Puede ser una tabla grande, leemos solo algunas columnas si es necesario
    df_habi = pd.read_excel(file_path)
    print(f'Total de registros cargados: {len(df_habi):,}')
    display(df_habi.head(3))
else:
    print('El archivo procesado no fue encontrado en la ruta esperada.')
```

## 📍 Territorial
**¿Qué localidades tienen peores indicadores? Ranking de vulnerabilidad**

```python
# Si existen variables de valor o nivel_indicador
if 'localidad' in df_habi.columns and 'valor_indicador' in df_habi.columns:
    # Promedio del indicador (Asegurarse de que mayor = peor, o menor = peor dependiendo del indicador)
    ranking_loc = df_habi.groupby('localidad')['valor_indicador'].mean().sort_values(ascending=False).head(15)
    
    plt.figure(figsize=(12,7))
    sns.barplot(x=ranking_loc.values, y=ranking_loc.index, palette='magma')
    plt.title('Promedio de Valor de Indicadores por Localidad')
    plt.xlabel('Valor Promedio del Indicador')
    plt.ylabel('Localidad')
    plt.show()
```

## 📊 Indicadores clave
**¿Cuáles indicadores son más críticos y cómo varían entre zonas?**

```python
if 'clase_indicador' in df_habi.columns:
    top_indicadores = df_habi['clase_indicador'].value_counts().head(8)
    plt.figure(figsize=(12,5))
    sns.barplot(x=top_indicadores.values, y=top_indicadores.index, palette='viridis')
    plt.title('Tipos de Indicadores Medidos con Frecuencia')
    plt.xlabel('Frecuencia de Registro')
    plt.show()
```

## 🧭 Distribución y Clusters
**¿Hay clusters de alta vulnerabilidad a nivel de UPZ?**

```python
if 'unidad_planeacion_zonal' in df_habi.columns:
    ranking_upz = df_habi['unidad_planeacion_zonal'].value_counts().head(10)
    print('Top 10 UPZs con mayor intensidad o frecuencia de fallas/indicadores:\n', ranking_upz)
```
