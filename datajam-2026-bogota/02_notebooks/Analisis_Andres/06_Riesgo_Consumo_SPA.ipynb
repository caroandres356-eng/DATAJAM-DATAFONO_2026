# 💊 6. Salud: Consumo SPA (FACTOR DE RIESGO)

👉 Identifica zonas y poblaciones con mayor riesgo de drogadicción al ser este un co-factor de las ideaciones suicidas.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

### Carga de Datos Preprocesados

```python
PROCESSED_DIR = '../../01_datos/processed'
file_path = os.path.join(PROCESSED_DIR, 'p_salud_consumo.xlsx')

if os.path.exists(file_path):
    df_spa = pd.read_excel(file_path)
    print(f'Total de registros cargados: {len(df_spa):,}')
    display(df_spa.head(3))
else:
    print('El archivo de Consumo SPA procesado no fue encontrado.')
```

## 📈 Tendencia
**¿Aumenta el consumo históricamente?**

```python
if 'ANO' in df_spa.columns:
    tendencia = df_spa['ANO'].value_counts().sort_index()
    plt.figure(figsize=(10,5))
    sns.lineplot(x=tendencia.index, y=tendencia.values, marker='D', color='purple', linewidth=2)
    plt.title('Curva Histórica de Casos de Consumo SPA')
    plt.xlabel('Año')
    plt.ylabel('Cantidad de Requerimientos')
    plt.xticks(tendencia.index)
    plt.show()
```

## 📍 Territorial
**¿Dónde es más alto el consumo / zonas críticas?**

```python
if 'NOMBRELOCALIDADRESIDENCIA' in df_spa.columns:
    ranking_spa = df_spa['NOMBRELOCALIDADRESIDENCIA'].value_counts().head(10)
    plt.figure(figsize=(12,6))
    sns.barplot(x=ranking_spa.values, y=ranking_spa.index, palette='inferno')
    plt.title('Top Localidades con Mayor Casos de Consumo SPA')
    plt.xlabel('Cantidad de Casos')
    plt.show()
```

## 📌 Contexto
**¿Dónde consumen (parques, vivienda, vía pública)?**

```python
cols_contexto = [col for col in df_spa.columns if 'SITIO' in col]
if cols_contexto:
    # Hacemos conteo de presencias (1, 'SI', etc)
    # Normalizamos el tipo de dato
    df_contexto = df_spa[cols_contexto].copy()
    for c in cols_contexto:
        df_contexto[c] = pd.to_numeric(df_contexto[c], errors='coerce').fillna(0)
        
    sumas_contexto = df_contexto.sum().sort_values(ascending=False)
    
    plt.figure(figsize=(12,6))
    sns.barplot(x=sumas_contexto.values, y=sumas_contexto.index, palette='YlOrBr_r')
    plt.title('Sitio Habitual de Consumo más recurrente')
    plt.xlabel('Veces Mencionado')
    plt.show()
```

## 👥 Perfil
**Edad / ciclo de vida y Sexo predominante**

```python
if 'CURSO_DE_VIDA' in df_spa.columns and 'SEXO' in df_spa.columns:
    plt.figure(figsize=(12,7))
    sns.countplot(data=df_spa, y='CURSO_DE_VIDA', hue='SEXO', palette='pastel', order=df_spa['CURSO_DE_VIDA'].value_counts().index)
    plt.title('Distribución de Consumidores por Curso de Vida y Sexo')
    plt.xlabel('Volumen de Casos')
    plt.ylabel('Curso de Vida')
    plt.show()
```
